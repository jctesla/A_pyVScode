```python
# sustraemos el Tocken y el id_grpChat (q es el id unico del grupo del chat, q se crea al crear el chabot
# de la cuenta de juan  carlos dergan)
import time
import requests
id_grpChat = -1
tocken = '6249959391:AAGko3-WyPRN6O1FYrRiRloM3ABULR3POaE'
url_tocken = f'https://api.telegram.org/bot{tocken}/getUpdates'
response = requests.get(url_tocken)
if response.status_code == 200:
  content_type = response.headers.get('Content-Type')
  print(content_type)                                     # es del tipo application/json
  if content_type == "application/json":
    data = response.json()                                # el tipo de dato devuelto por elservidor es josn
    # print(data,"\n")
    # for k,v in data.items():
    #  print("k=",k,"   v=",v)
    print("Consulta (True/False) :", data['ok'])
    lst_result = data['result']
    id_grpChat = lst_result[0]['message']['chat']['id']
    print("Id del Grupo de Chat :", id_grpChat)
      
      
# print(data)      
# {'ok': True, 'result': [{'update_id': 127682907, 'message': {'message_id': 4, 'from': {'id': 1360215944, 'is_bot': False, 'first_name': 'jc', 'last_name': 'dergan f.', 'username': 'J_Carlos_Dergan'}, 'chat': {'id': -884531886, 'title': 'Test_miTelegram', 'type': 'group', 'all_members_are_administrators': True}, 'date': 1683332908, 'text': '/my_id @user_jc_bot', 'entities': [{'offset': 0, 'length': 6, 'type': 'bot_command'}, {'offset': 7, 'length': 12, 'type': 'mention'}]}}]}      
```
